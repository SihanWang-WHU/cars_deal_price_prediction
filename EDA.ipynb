#### 无界约翰逊分布（Johnson SU Distribution）
无界约翰逊分布是一种灵活的分布类型，适用于各种形状的数据，特别是那些不符合标准正态分布或对数正态分布的数据。它可以处理具有不对称（偏态）和/或不规则峰态的数据集。
##### 灵活性：无界约翰逊分布是一种高度适应性的分布，能够拟合许多不符合简单分布模式的数据。
##### 参数：该分布通过两个形状参数、一个位置参数和一个尺度参数来定义，使其能够适应各种数据形状。
##### 应用场景：常用于金融和环境数据分析，特别是在数据展示出明显的偏态或峰态时。

有界约翰逊分布：
约翰逊SB分布适用于数据集有明显的下界和上界的情况。这意味着所有数据值都位于某个固定区间内，例如[0, 1]

#### 对数正态分布（Log-normal Distribution）
对数正态分布是一种连续概率分布，其特点是数据集的自然对数是正态分布的。这种分布通常用于描述那些不能为负且呈现正偏态（右偏）的数据。
##### 特性：如果数据遵循对数正态分布，那么它们的对数将遵循正态分布。
##### 应用场景：常见于描述自然界和经济活动中的各种现象，如收入分配、城市人口规模等。
##### 处理偏态数据：当原始数据呈现显著的正偏态时，对数转换常被用于稳定方差和使数据更接近正态分布。

补充知识<a href="https://www.cnblogs.com/cgmcoding/p/13384535.html" target="_blank">sns.distplot()的用法</a>

```python
## 1) 总体分布概况（无界约翰逊分布等）
import scipy.stats as st
y = train_data['price']
plt.figure(1); plt.title('Johnson SU')
# 使用 seaborn 的 distplot 函数绘制 y 的直方图，不绘制核密度估计（kde），并拟合 Johnson SU 分布
sns.distplot(y, kde=False, fit=st.johnsonsu)
plt.figure(2); plt.title('Normal')
# 绘制 y 的直方图，并拟合正态分布
sns.distplot(y, kde=False, fit=st.norm)
plt.figure(3); plt.title('Log Normal')
# 绘制 y 的直方图，并拟合对数正态分布
sns.distplot(y, kde=False, fit=st.lognorm)
```

```python
## 2) 查看skewness and kurtosis
sns.distplot(train_data['price']);
print("Skewness: %f" % train_data['price'].skew())
print("Kurtosis: %f" % train_data['price'].kurt())
```
